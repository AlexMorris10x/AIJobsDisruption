```python

def splitWordsBySpace(list):
    newList = []
    for word in list:
        newList.append(word.split(' '))
    return newList

def makeListOfWordsFromColumn(df, column):
    list = []
    for words in df[column]:
        words = words.split(' ')
        for word in words:
            list.append(word)
    return list

occupationList = makeListOfWordsFromColumn(noFormalEducationLowWage, 'OCCUPATION')

occupationList

```

```python
%pip install wordcloud
from wordcloud import WordCloud 

def makeWordCloud(list, title):
    wordcloud = WordCloud(width=800, height=800, max_font_size=200).generate(' '.join(list))
    plt.figure(figsize=(10,10))
    plt.title(title)
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.show()

makeWordCloud(occupationList, 'Lowest wage occupations & least qualified occupations')
```

```python
def removeWordsFromList(list, censoredWords):
    tmpList = []
    for word in list:
        if word in censoredWords:
            continue
        else:
            tmpList.append(word)
    return tmpList

cleanedOccupationList =  removeWordsFromList(occupationList, ['and', 'of', 'other', 'food','related', 'workers'])
cleanedOccupationList
```

```python
def frequencyDistributionFromList(list):
    from collections import Counter
    word_count = Counter(list)
    return word_count

frequencyDistributionCounter = frequencyDistributionFromList(cleanedOccupationList).most_common(50)

# frequencyDistributionCounter
# def sortCounter(counter):
#     return counter.most_common()

# sortedFrequencyDistribution = sortCounter(frequencyDistributionCounter)
# sortedFrequencyDistribution

def tupleToDictionary(tuple):
    dictionary = {}
    for key, value in tuple:
        dictionary[key] = value
    return dictionary

frequencyDistributionCounter = tupleToDictionary(frequencyDistributionCounter)
frequencyDistributionCounter
```

```python
def makeFrequencyDistributionChart(list, title):
    plt.figure(figsize=(10,10))
    plt.title(title)
    plt.bar(list.keys(), list.values())
    plt.xticks(rotation=90)
    plt.show()

makeFrequencyDistributionChart(frequencyDistributionCounter, 'Frequency distribution of occupations')
```
