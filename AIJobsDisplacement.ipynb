```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# functions 
def openCSV(fileName):
    df = pd.read_csv(fileName)
    return df

def dropNA(df):
    df = df.dropna()
    return df

def deleteEveryOtherRow(df):
    df = df.iloc[1::2]
    return df

def saveFile(df, fileName):
    df.to_csv(fileName, index=False)

def renameColumns(df, array):
    df.columns = [array]
    return df
```

```python
def findListDifference(list1, list2):
    difference = []
    for item in list1:
        if item not in list2:
            difference.append(item)
    return difference

# findListDifference(SectorTotalJobsList, SectorGDPList)
findListDifference(SectorGDPList, SectorTotalJobsList)
```

```python
SectorTotalJobs = SectorTotalJobs.sort_values(by=['Employment, 2020'])


def makePieChart(dataframe, column, title):
    plt.figure(figsize=(8,8))
    plt.title(title)
    plt.pie(dataframe[column], labels=dataframe['Industry Sector'], autopct='%1.1f%%')
    plt.axis('equal')
    plt.show()

makePieChart(SectorTotalJobs, 'Employment, 2020', 'Industry Sector')
```
